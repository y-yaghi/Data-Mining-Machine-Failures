```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('updated_balanced_dataset.csv')
```

```python
df.head()
```

```python
column1 = 'Rotational speed [rpm]'
column2 = 'Torque [Nm]'
df = df.drop(columns=[column1, column2])
df.rename(columns={'Machine Failure': 'All_MF'}, inplace=True)
df.rename(columns={'Machine Failure(2)': 'Balanced_MF'}, inplace=True)
df.to_csv("updated_balanced_dataset(2).csv", index=False) 
```

```python
feature_columns = ["Tool wear [min]", "Horse Power"]
scaler = MinMaxScaler()
df[feature_columns] = scaler.fit_transform(df[feature_columns])
df.to_csv("normalized_balanced_dataset(2).csv", index=False)
```
